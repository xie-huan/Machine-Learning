```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,10,100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
siny = y.copy()
cosy = np.cos(x)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--")
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--")
plt.xlim(0,10)
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--")
plt.axis([0,100,0,100])
```

```python
plt.plot(x,siny)
plt.plot(x,cosy,color="red",linestyle="--")
plt.xlabel("xlabel")
```

```python
plt.plot(x,siny,label="sinx")
plt.plot(x,cosy,color="red",linestyle="--",label="cosx")
plt.xlabel("xlabel")
plt.title("hello world")
plt.legend()
```

### 散点图

```python
plt.scatter(x,siny)
plt.scatter(x,cosy)
plt.show()
```
